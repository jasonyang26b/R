```r
install.packages('sandwich')
install.packages('lmtest')

library(tidyverse)
library(sandwich)
library(lmtest)
```

```r
d <- data.frame(
  x_1 = runif(1000), # random draw from uniform distribution
  x_2 = runif(1000),
  x_3 = runif(1000),
  x_4 = runif(1000)) %>%
    mutate(
    y = -1 + 1 * x_1 + 2 * x_2 + 0 * x_3 + 0 * x_4 + rnorm(1000)
    )
```

```r
head(d)
```

```r
model_short <- lm(y ~ 1, data = d)
model_long_a <- lm(y ~ 1 + x_1 + x_2, data = d)
model_long_b <- lm( y ~ 1 + x_1 + x_2 + x_3 + x_4, data = d)
```

```r
model_short
```

```r
model_long_a
```

# **anova**

```r
anova(model_long_a, model_short, test='F')
```

```r
anova(model_long_b, model_short, test='F')
```

```r
anova(model_long_b, model_long_a, test='F')
```

# **coeftest & vcovHC**
## **Single Coefficient**



```r
coeftest(model_short) # this is the default classical variance-covariance matrix
```

```r
vcovHC(model_short)
```

```r
coeftest(model_short, vcovHC(model_short))
# coeftest: from 'lmtest', hypothesis tests for regression coefficients.
# by default, it reports:
  # coefficient estimate
  # standard error
  # t_statistic
  # p-value

# vcovHC() from 'sandwich', a heteroskedasticity-consistent (robust) variance-covariance matrix.
# instead of classical SE, it uses the robust sandwich estimator.
```

## **Multiple coefficients**

```r
vcovHC(model_long_a)
coeftest(model_long_a, vcovHC(model_long_a))
```

```r
vcovHC(model_long_b)
coeftest(model_long_b, vcovHC(model_long_b))
```
